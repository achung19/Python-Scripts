```python
import re
dates = master_log['unit_name']
date_arr = dates.unique()
# print("date_arr", date_arr)

temp = []
for i in date_arr:
    unit_removed = re.sub(r"_T.*$","",i)
    temp.append(unit_removed)
    
temp = np.unique(temp)

unique_dates = []
for i in temp:
    for j in date_arr:
        if i in j:
            unique_dates.append(j)
            break    

# print("unique_dates", unique_dates)

cl4_s = re.compile("Cl4_.*")
cl4_list = list(filter(cl4_s.match, unique_dates))

cl5_s = re.compile("Cl5_")
cl5_list = list(filter(cl5_s.match, unique_dates))

cl6_s = re.compile("Cl6_")
cl6_list = list(filter(cl6_s.match, unique_dates))
```

```python
# start at trial 5, take a 10 trial window and calculate the ratio

def frac_calc(response_type, data_frame):
    i = 0.
    count = 0.
    fracArr = [0.] * (data_frame.shape[0]+1)
    while (i < data_frame.shape[0]):
        if (data_frame['Stim/Block/Response'].iloc[int(i)] == response_type[0] or data_frame['Stim/Block/Response'].iloc[int(i)] == response_type[1]):
            count += 1
        i += 1
        fracArr[int(i)] = count/i
    return fracArr
```

```python
# General Function
# args: mouse_date
def make_plot (mouse_date):
    unit_log = master_log[master_log['unit_name']== mouse_date]

    MissFrac = frac_calc(list({'SomMiss', 'VisMiss'}), unit_log)
    FAFrac = frac_calc(list({'SomFA', 'VisFA'}), unit_log)
    CRFrac = frac_calc(list({'SomCR', 'VisCR'}), unit_log)
    HitFrac = frac_calc(list({'SomHit', 'VisHit'}), unit_log)

    line1 = [0.] * (unit_log.shape[0]+1)
    line2 = [0.] * (unit_log.shape[0]+1)
    line3 = [0.] * (unit_log.shape[0]+1)

    for i in range(unit_log.shape[0]):
        line1[i] = HitFrac[i]
        line2[i] = HitFrac[i]+CRFrac[i]
        line3[i] = HitFrac[i]+CRFrac[i]+MissFrac[i]

    x = list(np.arange(unit_log.shape[0]+1, dtype=float))
    y_base = [0.] * (unit_log.shape[0]+1)
    y_max = [1.] * (unit_log.shape[0]+1)
    
    fig1 = plt.figure()
    ax1 = fig1.subplots()    
    ax1.fill_between(x, y_base, line1, color = '#9AD4D6', label = 'hits')
    ax1.fill_between(x, line1, line2, color = '#0FA3B1', label = 'crs')
    ax1.fill_between(x, line2, line3, color = '#F9F7F3', label = 'misses')
    ax1.fill_between(x, line3, y_max, color = '#D2D7DF', label = 'fas')

    plt.xlim(0,unit_log.shape[0])
    plt.ylim(0,1)

    r1 = Rectangle((0, 0), 1, 1, color = '#9AD4D6', ec = 'black')
    r2 = Rectangle((0, 0), 1, 1, color = '#0FA3B1', ec = 'black')
    r3 = Rectangle((0, 0), 1, 1, color = '#F9F7F3', ec = 'black')
    r4 = Rectangle((0, 0), 1, 1, color = '#D2D7DF', ec = 'black')

    plt.xlabel('Trials')
    plt.ylabel('Ratio of Response')
    plt.title('Running Ratio of Responses')

    legend1 = plt.legend([r1, r2, r3, r4], ["hits", "crs", "misses", "fas"], loc='lower right')
    
    return fig1
```

```python
# Claustrum 4 Behavior Analysis
for n in cl4_list:
    make_plot(n).savefig("test.pdf")
```

```python
# Claustrum 5 Behavior Analysis
for n in cl5_list:
    make_plot(n)
```
