```python
import re
dates = master_log['unit_name']
date_arr = dates.unique()
# print("date_arr", date_arr)

temp = []
for i in date_arr:
    unit_removed = re.sub(r"_T.*$","",i)
    temp.append(unit_removed)
    
temp = np.unique(temp)

unique_dates = []
for i in temp:
    for j in date_arr:
        if i in j:
            
            
            unique_dates.append(j)
            break    

# print("unique_dates", unique_dates)

cl4_s = re.compile("Cl4_.*")
cl4_list = list(filter(cl4_s.match, unique_dates))

cl5_s = re.compile("Cl5_")
cl5_list = list(filter(cl5_s.match, unique_dates))

cl6_s = re.compile("Cl6_")
cl6_list = list(filter(cl6_s.match, unique_dates))
```

```python
# cumulatively calculate rate for trials before reaching window_size
# move along window and calculate rate based off of previous window_size trials

def frac_calc(data_frame, window_size):
    hit_arr = np.zeros(shape = data_frame.shape[0]+1)
    miss_arr = np.zeros(shape = data_frame.shape[0]+1)
    CR_arr = np.zeros(shape = data_frame.shape[0]+1)
    FA_arr = np.zeros(shape = data_frame.shape[0]+1) 
    
    # index holds current calculating index, window_ptr holds index of other end of sliding window
    # num_X holds the number of responses of each type in the window
    index, window_ptr = (0,)*2
    num_Hit, num_Miss, num_CR, num_FA = (0.,)*4
    
    # cumulatively calculate rate for first few trials before reaching max window size
    while (index < window_size):
        if (data_frame['Stim/Block/Response'].iloc[int(index)] == 'SomHit' or data_frame['Stim/Block/Response'].iloc[int(index)] == 'VisHit'):
            num_Hit += 1
        elif (data_frame['Stim/Block/Response'].iloc[int(index)] == 'SomMiss' or data_frame['Stim/Block/Response'].iloc[int(index)] == 'VisMiss'):
            num_Miss += 1
        elif (data_frame['Stim/Block/Response'].iloc[int(index)] == 'SomCR' or data_frame['Stim/Block/Response'].iloc[int(index)] == 'VisCR'):
            num_CR += 1
        else:
            num_FA += 1
        index += 1
        hit_arr[index] = num_Hit/index
        miss_arr[index] = num_Miss/index
        CR_arr[index] = num_CR/index
        FA_arr[index] = num_FA/index
        
    # calculate along sliding window, subtracting when moving end of window forward and adding when moving window start forward
    while (index < data_frame.shape[0]):
        if (data_frame['Stim/Block/Response'].iloc[int(index)] == 'SomHit' or data_frame['Stim/Block/Response'].iloc[int(index)] == 'VisHit'):
            num_Hit += 1
        elif (data_frame['Stim/Block/Response'].iloc[int(index)] == 'SomMiss' or data_frame['Stim/Block/Response'].iloc[int(index)] == 'VisMiss'):
            num_Miss += 1
        elif (data_frame['Stim/Block/Response'].iloc[int(index)] == 'SomCR' or data_frame['Stim/Block/Response'].iloc[int(index)] == 'VisCR'):
            num_CR += 1
        else:
            num_FA += 1
            
        if (data_frame['Stim/Block/Response'].iloc[int(window_ptr)] == 'SomHit' or data_frame['Stim/Block/Response'].iloc[int(window_ptr)] == 'VisHit'):
            num_Hit += -1
        elif (data_frame['Stim/Block/Response'].iloc[int(window_ptr)] == 'SomMiss' or data_frame['Stim/Block/Response'].iloc[int(window_ptr)] == 'VisMiss'):
            num_Miss += -1
        elif (data_frame['Stim/Block/Response'].iloc[int(window_ptr)] == 'SomCR' or data_frame['Stim/Block/Response'].iloc[int(window_ptr)] == 'VisCR'):
            num_CR += -1
        else:
            num_FA += -1
            
        index += 1
        window_ptr += 1
        hit_arr[index] = num_Hit/window_size
        miss_arr[index] = num_Miss/window_size
        CR_arr[index] = num_CR/window_size
        FA_arr[index] = num_FA/window_size

    return hit_arr, miss_arr, CR_arr, FA_arr

```

```python
# General Function to create a plot for a mouse_date, with window_size
# creates a plot at fig at subplot location loc, with labels on or off
def make_plot (mouse_date, window_size, fig, loc = None, label = None):
    # unit_log for specific tetrode recording unit on a day
    unit_log = master_log[master_log['unit_name']== mouse_date]
    
    # calling rate function to get rates of each response type
    Hit_Frac, Miss_Frac, CR_Frac, FA_Frac = frac_calc(unit_log, window_size)
    
    # create line boundaries for fill_between
    x = list(np.arange(unit_log.shape[0]+1, dtype=float))
    y_base = [0.] * (unit_log.shape[0]+1)
    y_max = [1.] * (unit_log.shape[0]+1)
    line1 = [0.] * (unit_log.shape[0]+1)
    line2 = [0.] * (unit_log.shape[0]+1)
    line3 = [0.] * (unit_log.shape[0]+1)
    for i in range(unit_log.shape[0]):
        line1[i] = Hit_Frac[i]
        line2[i] = Hit_Frac[i]+CR_Frac[i]
        line3[i] = Hit_Frac[i]+CR_Frac[i]+Miss_Frac[i]
    
    # plot each figure using pretty colors
    subplot = fig.add_subplot(loc[0], loc[1], loc[2])
    subplot.fill_between(x, y_base, line1, color = '#9AD4D6', label = 'hits')
    subplot.fill_between(x, line1, line2, color = '#0FA3B1', label = 'crs')
    subplot.fill_between(x, line2, line3, color = '#F9F7F3', label = 'misses')
    subplot.fill_between(x, line3, y_max, color = '#D2D7DF', label = 'fas')

    # cleaning up edges of the plots
    plt.xlim(0,unit_log.shape[0])
    plt.ylim(0,1)

    # creating legend rectangles
    r1 = Rectangle((0, 0), 1, 1, color = '#9AD4D6', ec = 'black')
    r2 = Rectangle((0, 0), 1, 1, color = '#0FA3B1', ec = 'black')
    r3 = Rectangle((0, 0), 1, 1, color = '#F9F7F3', ec = 'black')
    r4 = Rectangle((0, 0), 1, 1, color = '#D2D7DF', ec = 'black')

    # labels, titles, axes, etc
    if (label):
        plt.xlabel('Trials')
        plt.ylabel('Ratio of Responses')
        plt.title('Running Ratio of Responses of ' + mouse_date)
        plt.legend([r1, r2, r3, r4], ["hits", "crs", "misses", "fas"], loc='lower right')
```

```python
# Prints 2 plots to a pdf page called pdf_name using data from mouse_data_list 
def print_plots(mouse_data_list, pdf_name):
    with PdfPages(pdf_name + ".pdf") as pdf:

        fig = plt.figure()
        fig.set_figheight(10)
        fig.set_figwidth(8)
        k=1
        for n in mouse_data_list:
            make_plot(n, 20, fig, loc = (2,1,k), label = True)
            k += 1
            if (k > 2):
                pdf.savefig(fig)
                fig = plt.figure()
                fig.set_figheight(10)
                fig.set_figwidth(8)
                k = 1

        if (k == 2):
            pdf.savefig(fig)
```
